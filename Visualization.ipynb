```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns # Import seaborn

# Load CSV
file_path1 = '/content/drive/MyDrive/data_har/Gyroscope rotation rate 2025-05-19 11-19-15.zip (Unzipped Files)/Raw Data.csv'  # Replace with your actual file path after download
df1 = pd.read_csv(file_path1)
file_path2 = '/content/drive/MyDrive/data_har/Gyroscope rotation rate 2025-05-19 11-22-21.zip (Unzipped Files)/Raw Data.csv'  # Replace with your actual file path after download
df2 = pd.read_csv(file_path2)
accel_path1 = '/content/drive/MyDrive/data_har/DOC-20250519-WA0001.zip (Unzipped Files)/Raw Data.csv'
accel_path2 = '/content/drive/MyDrive/data_har/DOC-20250519-WA0002.zip (Unzipped Files)/Raw Data.csv'
accel_1 = pd.read_csv(accel_path1)
accel_2 = pd.read_csv(accel_path2)

# Check the column names of accel_1 and accel_2 to verify
#print("Columns in accel_1:", accel_1.columns) # Added descriptive text
#print("Columns in accel_2:", accel_2.columns) # Added descriptive text

# Buat subplot 2 baris dan 4 kolom
fig, axes = plt.subplots(4, 4, figsize=(20, 20))

# Baris 1
sns.lineplot(data=df1, x='Time (s)', y='Gyroscope x (rad/s)', ax=axes[0][0])
axes[0][0].set_title('Gyroscope x over Time')

sns.lineplot(data=df1, x='Time (s)', y='Gyroscope y (rad/s)', ax=axes[0][1])
axes[0][1].set_title('Gyroscope y over Time')

sns.lineplot(data=df1, x='Time (s)', y='Gyroscope z (rad/s)', ax=axes[0][2])
axes[0][2].set_title('Gyroscope z over Time')

sns.lineplot(data=df1, x='Time (s)', y='Absolute (rad/s)', ax=axes[0][3])
axes[0][3].set_title('Absolute Gyroscope over Time')

# Baris 2 – tambahkan data lain sesuai kebutuhan
sns.lineplot(data=df2, x='Time (s)', y='Gyroscope x (rad/s)', ax=axes[1][0])
axes[1][0].set_title('Gyroscope x over Time')

sns.lineplot(data=df2, x='Time (s)', y='Gyroscope y (rad/s)', ax=axes[1][1])
axes[1][1].set_title('Gyroscope y over Time')

sns.lineplot(data=df2, x='Time (s)', y='Gyroscope z (rad/s)', ax=axes[1][2])
axes[1][2].set_title('Gyroscope z over Time')

sns.lineplot(data=df2, x='Time (s)', y='Absolute (rad/s)', ax=axes[1][3])
axes[1][3].set_title('Absolute Gyroscope over Time')

# Baris 3 – tambahkan data lain sesuai kebutuhan
sns.lineplot(data=accel_1, x='Time (s)', y='Acceleration x (m/s^2)', ax=axes[2][0])
axes[2][0].set_title('Acceleration x over Time') # Corrected index to axes[2][0]

sns.lineplot(data=accel_1, x='Time (s)', y='Acceleration y (m/s^2)', ax=axes[2][1])
axes[2][1].set_title('Acceleration y over Time') # Corrected index to axes[2][1]

sns.lineplot(data=accel_1, x='Time (s)', y='Acceleration z (m/s^2)', ax=axes[2][2])
axes[2][2].set_title('Acceleration z over Time') # Corrected index to axes[2][2]

sns.lineplot(data=accel_1, x='Time (s)', y='Absolute acceleration (m/s^2)', ax=axes[2][3])
axes[2][3].set_title('Absolute Acceleration over Time') # Corrected index to axes[2][3] and title

# Baris 4 – tambahkan data lain sesuai kebutuhan
sns.lineplot(data=accel_2, x='Time (s)', y='Acceleration x (m/s^2)', ax=axes[3][0])
axes[3][0].set_title('Acceleration x over Time') # Corrected index to axes[3][0]

sns.lineplot(data=accel_2, x='Time (s)', y='Acceleration y (m/s^2)', ax=axes[3][1])
axes[3][1].set_title('Acceleration y over Time') # Corrected index to axes[3][1]

sns.lineplot(data=accel_2, x='Time (s)', y='Acceleration z (m/s^2)', ax=axes[3][2])
axes[3][2].set_title('Acceleration z over Time') # Corrected index to ax

sns.lineplot(data=accel_2, x='Time (s)', y='Absolute acceleration (m/s^2)', ax=axes[3][3])
axes[3][3].set_title('Absolute Acceleration over Time') # Corrected index to axes[3][3] and title


# Rapikan layout
plt.tight_layout()
plt.show()
```
