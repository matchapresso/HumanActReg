```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns # Import seaborn
```

```python
def plot_sensor_data(gyro_path1, gyro_path2, accel_path1, accel_path2, save_path=None, title_prefix=''):
    """
    Visualisasi subplot 4x4 untuk dua file gyroscope dan dua file accelerometer.
    Kompatibel untuk environment seperti Kaggle.
    
    Parameters:
        gyro_path1 (str): Path ke file gyroscope pertama
        gyro_path2 (str): Path ke file gyroscope kedua
        accel_path1 (str): Path ke file akselerometer pertama
        accel_path2 (str): Path ke file akselerometer kedua
        save_path (str, optional): Jika ingin menyimpan hasil plot ke file PNG
        title_prefix (str, optional): Tambahan teks di awal setiap judul plot
    """
    # Load data
    df1 = pd.read_csv(gyro_path1)
    df2 = pd.read_csv(gyro_path2)
    accel_1 = pd.read_csv(accel_path1)
    accel_2 = pd.read_csv(accel_path2)

    # Buat figure dan axes 4x4
    fig, axes = plt.subplots(4, 3, figsize=(20, 20))

    # List kolom yang akan diplot
    gyro_cols = ['x', 'y', 'z']
    accel_cols = ['x', 'y', 'z']

    # Plot gyroscope file 1
    for i, col in enumerate(gyro_cols):
        sns.lineplot(data=df1, x='time', y=col, ax=axes[0][i])
        axes[0][i].set_title(f'{title_prefix} {col} (Gyro 1)')

    # Plot gyroscope file 2
    for i, col in enumerate(gyro_cols):
        sns.lineplot(data=df2, x='time', y=col, ax=axes[1][i])
        axes[1][i].set_title(f'{title_prefix} {col} (Gyro 2)')

    # Plot accelerometer file 1
    for i, col in enumerate(accel_cols):
        sns.lineplot(data=accel_1, x='time', y=col, ax=axes[2][i])
        axes[2][i].set_title(f'{title_prefix} {col} (Accel 1)')

    # Plot accelerometer file 2
    for i, col in enumerate(accel_cols):
        sns.lineplot(data=accel_2, x='time', y=col, ax=axes[3][i])
        axes[3][i].set_title(f'{title_prefix} {col} (Accel 2)')

    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path)
        print(f'Plot saved to {save_path}')
    else:
        plt.show()
```

```python
plot_sensor_data(
    '/kaggle/input/imu-data-human-activity/imu/Areta_1/Gyroscope.csv',
    '/kaggle/input/imu-data-human-activity/imu/Areta_2/Gyroscope.csv',
    '/kaggle/input/imu-data-human-activity/imu/Areta_1/Accelerometer.csv',
    '/kaggle/input/imu-data-human-activity/imu/Areta_2/Accelerometer.csv',
    save_path= None,  # Atau None untuk langsung tampil
    title_prefix='HAR A'
)
```
